```python
# Optimized disjoint check

set1 = input("Enter first set elements: ").split()
set2 = input("Enter second set elements: ").split()

found = False

for i in set1:
    for j in set2:
        if i == j:
            found = True
            break

    if found:
        break

if found:
    print("Sets are NOT Disjoint.")
else:
    print("Sets are Disjoint.")
```
